## Normalization of the paper

```python
def normal(X):
    for column in X.columns:
        minimum = np.min(X[column])
        m = np.mean(X[column])
        s = np.std(X[column])
        if(minimum>0):
            X[column] = X[column]-m+1
        else:
            X[column] = (X[column]-m)/s
    return X
```

```python
X_train_prepared = normal(X_train)
X_test_prepared = normal(X_test)
```
